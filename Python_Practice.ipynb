```python
def fibonacci(n):
    if n == 0:
        return []
    elif n == 1:
        return [0]
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i - 1] + fib[i - 2])
    return fib

print(fibonacci(5))  # Output: [0, 1, 1, 2, 3]
```

```python
from abc import ABC, abstractmethod

# Abstract class
class Vehicle(ABC):

    @abstractmethod
    def start_engine(self):
      print("hii")
      pass  # Must be implemented by child classes

    @abstractmethod
    def stop_engine(self):
        pass

# Concrete class implementing abstraction
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started with a key or button.")

    def stop_engine(self):
        print("Car engine stopped.")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine started with a kick or button.")

    def stop_engine(self):
        print("Bike engine stopped.")

# Using abstraction
vehicle1 = Car()
vehicle1.start_engine()
vehicle1.stop_engine()

vehicle2 = Bike()
vehicle2.start_engine()
vehicle2.stop_engine()
```

```python
class Car:
    def start_engine(self):
        print("Car engine- key .")

class Bike:
    def start_engine(self):
        print("Bike engine kick.")

class Airplane:
    def start_engine(self):
        print("Airplane engine pilot controls.")

# Function using polymorphism
def start_vehicle(vehicle):
    vehicle.start_engine()  # Works for Car, Bike, Airplane

# Test
start_vehicle(Car())       # Car engine started with a key
start_vehicle(Bike())      # Bike engine started with a kick
start_vehicle(Airplane())  # Airplane engine pilot controls.
```

```python
# 0  1  2
#10 20 30 ---0
#40 50 60 ---1
arr=np.array([[10,20,30],[40,50,60]])
print(arr)
print(arr[1,2])#accessing element
print(arr[0])#slicing row 0
print(arr[:,2])#slicing column 1
print(arr[-1,-1]) #negative indexing
print(arr[::-1])
```

```python
data=np.array([[1,2,3],[4,5,6]])
print(data)
print(np.sum(data)) #sum all
print(np.sum(data,axis=0))#sum columns
print(np.sum(data,axis=1))#sum rows
print(np.mean(data,axis=0))#mean col
print(np.mean(data,axis=1))#mean row
print(np.max(data))
print(np.min(data))
```
